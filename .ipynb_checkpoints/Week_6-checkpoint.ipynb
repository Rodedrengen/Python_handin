```python
#1.1 and 1.2
import numpy as np
import pandas as pd

%matplotlib notebook
import matplotlib.pyplot as plt

url = "https://api.statbank.dk/v1/data/FOLK1A/CSV?valuePresentation=CodeAndValue&delimiter=Semicolon&ALDER=%3E17%3C90&CIVILSTAND=*&K%C3%98N=2%2C1&OMR%C3%85DE=000&Tid=*"
newUrl = "https://api.statbank.dk/v1/data/FOLK1A/CSV?valuePresentation=CodeAndValue&delimiter=Semicolon&K%C3%98N=TOT&OMR%C3%85DE=000&ALDER=IALT&CIVILSTAND=TOT&Tid=2021K1"
myDataFrame = pd.read_csv(url, sep=';')
smallDataFrame = pd.read_csv(newUrl, sep=';')

totalAmountOfDanes = smallDataFrame["INDHOLD"]

times = ('2008K1 2008K1',
        '2009K1 2009K1',
        '2010K1 2010K1',
        '2011K1 2011K1',
        '2012K1 2012K1',
        '2013K1 2013K1',
        '2014K1 2014K1',
        '2015K1 2015K1',
        '2016K1 2016K1',
        '2017K1 2017K1',
        '2018K1 2018K1',
        '2019K1 2019K1',
        '2020K1 2020K1',
        '2021K1 2021K1')
mydict = {}

for t in times:
    mask = ((myDataFrame['CIVILSTAND'] == 'F Fraskilt') & (myDataFrame['TID'] == t))
    mydict[t] = myDataFrame[mask]["INDHOLD"].sum()
    
values = mydict.values()
keys = mydict.keys()

plt.bar(keys,values,width=0.5, align='center')
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
```
